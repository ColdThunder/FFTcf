```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
from astropy.table import Table


plt.style.use('presentation')



```

```python



filename='../output/2pcf.txt'
print(filename)
t = Table.read(filename,format='ascii')
YYkk=t['col1']
YYp0=t['col3']
YYp2=t['col5']
YYp4=t['col7']
YYcc=t['col9']

#plot
plt.figure(1)
plt.subplot(111)

plt.axis([0,200,-15,50])

plt.plot(YYkk,YYkk**2*YYp0,'k-',lw=3,label='$\\xi_0$')
plt.plot(YYkk,YYkk**2*YYp2,'r-',lw=3,label='$\\xi_2$')
plt.plot(YYkk,YYkk**2*YYp4,'b-',lw=3,label='$\\xi_4$')

plt.xscale('linear')
plt.yscale('linear')
plt.xlabel('$r$')
plt.ylabel('$r^2\\xi(r)$')
#plt.title('RecAni-'+rr)
plt.grid(True)

#plt.legend(loc='upper right',fontsize=12,ncol=2)

plt.savefig('xi.pdf',bbox_inches='tight')
plt.show()


```
